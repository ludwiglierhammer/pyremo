# Physics

pyremo contains a simple physics package with common algorithms that were used in a number of legacy fortran programs in and round REMO. For convenience, these algorithms are now implemented in the `pyremo.physics` package. Typical reacurring tasks are, e.g., the derivation relative (`RELHUM`) and specific humidity (`QD`) or liquid water content (`QW`) using temperature and pressure.

```python
from pyremo import physics
from pyremo import data
```

```python
ds = data.example_output()
```

```python
relhum = physics.relative_humidity(ds.T, ds.QD, ds.PS, ds.hyai, ds.hybi)
```
